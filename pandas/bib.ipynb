## Join DataFrame

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.join.html

```python
authors.join(authors_books, lsuffix='_caller', rsuffix='_other')
```

```python
authors.set_index('id').join(authors_books.set_index('id_author'))
```

```python
authors.join(authors_books.set_index('id_author'), on='id')
```

```python
# New in version 1.5.0.
authors.join(authors_books.set_index('id_author'), on='id', validate='m:1')
```

```python
authors.join(authors_books.set_index('id_author'), on='id').join(books.set_index('id_book'), on='id_book')
```
